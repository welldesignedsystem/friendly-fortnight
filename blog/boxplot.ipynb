### Box plots ‚≠ê - immediate visual of IQR outliers

# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
group_a = np.random.normal(100, 15, 500)
group_b = np.random.normal(110, 20, 500)
group_c = np.random.normal(95, 10, 500)

data_groups = [group_a, group_b, group_c]
# Compare multiple distributions
group_a = np.random.normal(100, 15, 500)
group_b = np.random.normal(110, 20, 500)
group_c = np.random.normal(95, 10, 500)

data_groups = [group_a, group_b, group_c]

plt.figure(figsize=(10, 6))
plt.boxplot(data_groups, labels=['Group A', 'Group B', 'Group C'])
plt.ylabel('Value')
plt.title('Comparing Multiple Distributions')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

# Seaborn

```python
import seaborn as sns
import pandas as pd

# Create DataFrame for easier plotting
df = pd.DataFrame({
    'Group': ['A']*500 + ['B']*500 + ['C']*500,
    'Value': np.concatenate([group_a, group_b, group_c])
})

# Basic seaborn box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Group', y='Value', palette='Set2')
plt.title('Box Plot with Seaborn')
plt.show()

# With individual data points
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Group', y='Value', palette='Set2')
sns.stripplot(data=df, x='Group', y='Value', color='black', alpha=0.3, size=2)
plt.title('Box Plot with Data Points Overlay')
plt.show()
```

```python
# Violin plot shows distribution shape along with box plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Group', y='Value', palette='Set2')
plt.title('Violin Plot - Box Plot + Distribution')
plt.show()

# Split violin plot for comparing categories
df['Category'] = np.random.choice(['X', 'Y'], len(df))
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Group', y='Value', hue='Category', split=True, palette='Set1')
plt.title('Split Violin Plot')
plt.show()
```

```python
data = np.random.normal(100, 15, 500)

# Calculate IQR and identify outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = data[(data < lower_bound) | (data > upper_bound)]

print(f"Q1: {Q1:.2f}")
print(f"Q3: {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Outlier bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
print(f"Number of outliers: {len(outliers)}")
print(f"Outlier values: {outliers[:10]}")  # Show first 10
```

```python
# Direct plotting from DataFrame
df_data = pd.DataFrame({
    'Score_A': np.random.normal(85, 10, 100),
    'Score_B': np.random.normal(78, 15, 100),
    'Score_C': np.random.normal(90, 8, 100)
})

# Pandas built-in box plot
df_data.boxplot(figsize=(10, 6))
plt.title('Box Plot from DataFrame')
plt.ylabel('Score')
plt.show()

# Or with seaborn for better styling
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_data, palette='viridis')
plt.title('Box Plot - Multiple Variables')
plt.ylabel('Score')
plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Box(
    y=group_a,
    name='Group A',
    boxmean='sd'  # Show mean and standard deviation
))

fig.add_trace(go.Box(
    y=group_b,
    name='Group B',
    boxmean='sd'
))

fig.add_trace(go.Box(
    y=group_c,
    name='Group C',
    boxmean='sd'
))

fig.update_layout(
    title='Interactive Box Plot with Plotly',
    yaxis_title='Value',
    showlegend=True
)

fig.show()
```


### Key Parameters to Remember

**Matplotlib `boxplot()` parameters:**
- `vert=True/False` - Vertical or horizontal orientation
- `patch_artist=True` - Enable color filling
- `notch=True` - Add notch for confidence interval around median
- `showmeans=True` - Display mean marker
- `showfliers=True` - Display outliers (default)
- `labels` - Labels for each box

**Seaborn `boxplot()` parameters:**
- `data` - DataFrame
- `x`, `y` - Column names
- `hue` - Grouping variable for colors
- `palette` - Color scheme
- `orient='v'/'h'` - Orientation

### When to Use Box Plots

‚úÖ **Use box plots when:**
- Comparing distributions across groups
- Identifying outliers quickly
- Showing median and quartiles
- Working with skewed data
- Need compact visualization of multiple distributions

‚ùå **Avoid box plots when:**
- You need to see exact distribution shape (use histogram/KDE instead)
- You have very small sample sizes (< 20 points)
- You need to show individual data points (use strip/swarm plot instead)
