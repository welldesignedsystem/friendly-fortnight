# Andrews Curves in Python: A Complete Guide



## What are Andrews Curves?

Andrews curves are a powerful visualization technique for multivariate data that transforms each high-dimensional observation into a smooth curve. This allows you to visualize patterns, clusters, and outliers in datasets with many features.

Analyzing high-dimensional data is challenging as no method exists to visualize more than three dimensions simultaneously. This is popularly renowned as the Curse of Dimensionality and refers to a scenario where we have “too many” features in our dataset.



**The Mathematical Formula:**

For a data point with features x₁, x₂, x₃, ..., xₙ, the Andrews curve is defined as:

```
f(t) = x₁/√2 + x₂·sin(t) + x₃·cos(t) + x₄·sin(2t) + x₅·cos(2t) + ...
```

where t ranges from -π to π.

## Why Use Andrews Curves?

- **Visualize high-dimensional data** in 2D
- **Identify clusters** - similar observations produce similar curves
- **Detect outliers** - unusual patterns stand out
- **Compare groups** - different classes show distinct curve patterns
- **Preserve structure** - maintains distance relationships from the original data

## Implementation in Python

### Method 1: Using Pandas (Easiest) ⭐

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from pandas.plotting import andrews_curves

# Load sample data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create Andrews curves
plt.figure(figsize=(12, 6))
andrews_curves(df, 'species', alpha=0.4, colormap='Set1')
plt.title('Andrews Curves - Iris Dataset')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

You see that the values of virginca and versicolor species cluster together or correlate quite well with each other, while setosa is more distinctly different from the other.
